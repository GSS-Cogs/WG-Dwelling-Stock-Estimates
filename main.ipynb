Dwelling stock estimates by local authority and tenure

```python
from gssutils import *
```

```python
urlist = ['https://statswales.gov.wales/v/FOel','https://statswales.gov.wales/v/FOeu',
          'https://statswales.gov.wales/v/FOev', 'https://statswales.gov.wales/v/FOfN',
          'https://statswales.gov.wales/v/FOfO', 'https://statswales.gov.wales/v/FOfP',
         'https://statswales.gov.wales/v/FOfQ','https://statswales.gov.wales/v/FOfR',
         'https://statswales.gov.wales/v/FOfS', 'https://statswales.gov.wales/v/FOfT',
         'https://statswales.gov.wales/v/FOfU','https://statswales.gov.wales/v/FOfV',
         'https://statswales.gov.wales/v/FOfW', 'https://statswales.gov.wales/v/FOfX',
         'https://statswales.gov.wales/v/FOfY','https://statswales.gov.wales/v/FOfZ',
         'https://statswales.gov.wales/v/FOfa']
```

```python
yearlist = ['gregorian-interval/2016-03-31T00:00:00/P1Y',
            'gregorian-interval/2015-03-31T00:00:00/P1Y',
            'gregorian-interval/2014-03-31T00:00:00/P1Y',
            'gregorian-interval/2013-03-31T00:00:00/P1Y',
            'gregorian-interval/2012-03-31T00:00:00/P1Y',
            'gregorian-interval/2011-03-31T00:00:00/P1Y',
            'gregorian-interval/2010-03-31T00:00:00/P1Y',
            'gregorian-interval/2009-03-31T00:00:00/P1Y',
            'gregorian-interval/2008-03-31T00:00:00/P1Y',
            'gregorian-interval/2007-03-31T00:00:00/P1Y',
            'gregorian-interval/2006-03-31T00:00:00/P1Y',
            'gregorian-interval/2005-03-31T00:00:00/P1Y',
            'gregorian-interval/2004-03-31T00:00:00/P1Y',
            'gregorian-interval/2003-03-31T00:00:00/P1Y',
            'gregorian-interval/2002-03-31T00:00:00/P1Y',
            'gregorian-interval/2001-03-31T00:00:00/P1Y',
            'gregorian-interval/2000-03-31T00:00:00/P1Y']
```

```python
table = pd.DataFrame()
for i in range(0,17):
    print(i)
    print(urlist[i])
    data = pd.read_html(urlist[i])
    df = data[0].iloc[14: , 0:7]
    df.columns = ['Area', 'local-authority', 'registered-social-landlord','owner-occupied-privately-rented',\
                 'owner-occupied','privately-rented','total']
    df['Period'] = yearlist[i]
    table = pd.concat([table, df])    
```

```python
table
```

```python
table = table[['Period','Area','local-authority','registered-social-landlord',
               'owner-occupied-privately-rented', 'owner-occupied','privately-rented', 'total']]
```

```python
table = pd.melt(table, ['Period','Area'], var_name = 'Tenure', value_name = 'Value')
```

```python
def user_perc(x):
    
    if 'r'in str(x)  :
        return 'revised'
    else:
        return 'original-estimate'
    
table['Revision'] = table.apply(lambda row: user_perc(row['Value']), axis = 1)
```

```python
table['Value'] = table['Value'].map(lambda x: x.lstrip('(r), '))
```

```python
table['Value'] = table['Value'].str.replace(',', '')
table['Value'] = table['Value'].str.replace('.', '')
```

```python
table['Geography'] = table['Area'].map(
    lambda x: {
        'Blaenau Gwent':'W06000019',
        'Bridgend':'W06000013',
        'Caerphilly':'W06000018',
        'Cardiff':'W06000015',
        'Carmarthenshire':'W06000010',
        'Ceredigion':'W06000008',
        'Conwy':'W06000003',
        'Denbighshire':'W06000004',
        'Flintshire':'W06000005',
        'Gwynedd':'W06000002',
        'Isle of Anglesey':'W06000001',
        'Merthyr Tydfil':'W06000024',
        'Monmouthshire':'W06000021',
        'Neath Port Talbot':'W06000012',
        'Newport':'W06000022',
        'Pembrokeshire':'W06000009',
        'Powys':'W06000023',
        'Rhondda Cynon Taf':'W06000016',
        'Swansea':'W06000011',
        'Torfaen':'W06000020',
        'Vale of Glamorgan':'W06000014',
        'Wales':'W92000004',
        'Wrexham':'W06000006'
        }.get(x, x))
```

```python
table['Unit'] = 'dwellings'
table['Measure Type'] = 'Count'
```

```python
table = table[['Geography','Period','Tenure','Value','Revision','Unit','Measure Type']]
```

```python
destinationFolder = Path('out')
destinationFolder.mkdir(exist_ok=True, parents=True)

table.to_csv(destinationFolder / ('observations.csv'), index = False)
```
